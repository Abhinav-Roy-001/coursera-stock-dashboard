```python
#Question 5: Plot Tesla Stock Graph
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create Tesla stock data manually (sample data)
tesla_stock = pd.DataFrame({
    "Date": pd.date_range(start="2017-01-01", periods=20, freq="Q"),
    "Close": [
        60, 70, 80, 85,
        100, 120, 150, 180,
        220, 250, 300, 350,
        400, 500, 600, 700,
        800, 900, 950, 1000
    ]
})

# Step 2: Tesla quarterly revenue (from Q2)
tesla_revenue = pd.DataFrame({
    "Date": [
        "2021-12-31", "2021-09-30", "2021-06-30", "2021-03-31",
        "2020-12-31", "2020-09-30", "2020-06-30", "2020-03-31",
        "2019-12-31", "2019-09-30", "2019-06-30", "2019-03-31",
        "2018-12-31", "2018-09-30", "2018-06-30", "2018-03-31",
        "2017-12-31", "2017-09-30", "2017-06-30", "2017-03-31"
    ],
    "Revenue": [
        "17719", "13757", "11958", "10389",
        "10744", "8771", "6036", "5985",
        "7384", "6303", "6350", "4541",
        "7212", "6824", "4002", "3409",
        "3288", "2985", "2790", "2696"
    ]
})

# Clean revenue data
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype(float)
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])

# Step 3: Plot both in a dashboard
fig, axes = plt.subplots(2, 1, figsize=(14,10))

# Stock price plot
axes[0].plot(tesla_stock["Date"], tesla_stock["Close"], label="Tesla Stock Price", color="blue")
axes[0].set_title("Tesla Stock Price (Sample Data)")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Price (USD)")
axes[0].legend()

# Revenue plot
axes[1].plot(tesla_revenue["Date"], tesla_revenue["Revenue"], marker="o", label="Tesla Quarterly Revenue", color="green")
axes[1].set_title("Tesla Quarterly Revenue")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Revenue (in millions USD)")
axes[1].legend()

plt.tight_layout()
plt.show()
```
